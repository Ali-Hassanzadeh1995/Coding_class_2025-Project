```python
from datetime import date, datetime
import yfinance as yf
import re

# --- Helper Functions ---

def valid_date_format(d):
    """Check date format YYYY-MM-DD using regex."""
    return bool(re.fullmatch(r"\d{4}-\d{2}-\d{2}", d))

def parse_date(d):
    """Convert string date to datetime object."""
    return datetime.strptime(d, "%Y-%m-%d").date()

# --- Today‚Äôs Date ---
today = date.today()
print(f"Note: Date format is YYYY-MM-DD (e.g., {today})")

# --- Find Earliest Available Dates ---
tickers = ['AAPL', 'MSFT', 'NVDA', 'GOOGL', 'IBM']
min_dates = []

for t in tickers:
    ticker_data = yf.Ticker(t).history(period="max")
    earliest = ticker_data.index.min().date()
    min_dates.append(earliest)
    print(f"The earliest valid date for {t} is {earliest}")

max_date = max(min_dates)
print(f"\nüëâ The latest of all minimum dates is {max_date}")

# --- Get Start Date ---
S_date_input = input("\nEnter Start Date (YYYY-MM-DD): ")

if valid_date_format(S_date_input):
    S_date = parse_date(S_date_input)
    if S_date < max_date:
        print(f"‚ùå Invalid: Start date must be >= {max_date}")
    else:
        print(f"‚úÖ Start date accepted: {S_date}")
else:
    print("‚ùå Invalid format! Please use YYYY-MM-DD")

# --- Get End Date ---
E_date_input = input("\nEnter End Date (YYYY-MM-DD): ")

if valid_date_format(E_date_input):
    E_date = parse_date(E_date_input)
    if E_date > today:
        print(f"‚ùå Invalid: End date must be <= {today}")
    else:
        print(f"‚úÖ End date accepted: {E_date}")
else:
    print("‚ùå Invalid format! Please use YYYY-MM-DD")
```
